```python
@app.callback(
    Output(component_id="fig_map", component_property='figure'),#, allow_duplicate=True),
    Input(component_id='slct-label', component_property='value')
    #prevent_initial_call = True
)
def updated_map(selected_label):
  print("Callback triggered!", flush=True)
  df_copy =df.copy()
  df_copy1 = df_copy.loc[df_copy['Has Online delivery']==selected_label]
  print(selected_label, flush=True)
  app.logger.debug(f"Selected label: {selected_label}")
  logging.info(f"--- Callback received value: {selected_label} ---")
  rest_map = px.scatter_geo(df_copy1,
                lat='Latitude',
                lon='Longitude',
                color='Has Online delivery',
                size='Aggregate rating',
                color_discrete_map={'No':'violet',
                                         'Yes':'lightgreen'},
                hover_name='City')
  rest_map.update_layout(
      margin=dict(
        l=0, r=0, t=0, b=0
       ),
        geo=dict(
        showframe=False,
        showcountries=True,
        fitbounds="locations"
        ),
        plot_bgcolor='red',
        showlegend = False,
        autosize=True,
        height=200,
        width=400
  )
  return rest_map
```
